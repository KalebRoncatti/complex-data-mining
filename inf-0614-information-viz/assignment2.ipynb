```python
import pandas
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
data = pandas.read_csv('periodic_table.csv', sep=',', encoding= 'unicode_escape').dropna()
data.head(5)

```

```python
subset = ['Type', 'Atomic Weight', 'Density', 'Specific Heat Capacity']
subset_without_string = ['Atomic Weight', 'Density', 'Specific Heat Capacity']

data_adapted = data[subset]
data_adapted[subset_without_string] = scaler.fit_transform(data_adapted[subset_without_string])


parallel_coordinates(data_adapted, 'Type')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```

```python
subset = ['Type', 'Atomic Weight', 'Boiling Point (K)']
subset_without_string = ['Atomic Weight', 'Boiling Point (K)']

data_adapted = data[subset]
data_adapted[subset_without_string] = scaler.fit_transform(data_adapted[subset_without_string])


parallel_coordinates(data_adapted, 'Type')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
```
